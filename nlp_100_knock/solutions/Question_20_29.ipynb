# 準備

```python
import random
import json
import requests
import re
import regex
import numpy as np
```

# 問題

## 20. JSONデータの読み込み

```python
json_path = 'Question_20_29/jawiki-country.json'
```

```python
with open(json_path, 'r', encoding='utf8') as f:
    articles = [json.loads(article_str) for article_str in f.readlines()]
```

```python
articles[1]
```

## 21. カテゴリ名を含む行を抽出

```python
pattern = re.compile(r'\[\[Category.*\]\]', re.MULTILINE)
category_lines = [pattern.findall(article['text']) for article in articles]
```

```python
category_lines[0]
```

## 22. カテゴリ名の抽出

```python
pattern = re.compile(r'\[\[Category:(?P<item>[^|]+?)(\|.+?)*?\]\]', re.MULTILINE)
categories = [[found[0] for found in pattern.findall(article['text'])] for article in articles]
```

```python
categories[0]
```

## 23. セクション構造

```python
pattern = re.compile(r'(?P<deco>={2,})\s+(?P<title>.*) \1', re.MULTILINE)
sections = [[f'{sect}({len(deco)})' for deco, sect in pattern.findall(article['text'])] for article in articles]
```

```python
sections[1]
```

## 24. ファイル参照の抽出

```python
pattern = re.compile(r'(\[\[(File|ファイル):.*\]\])', re.MULTILINE)
file_lines = [[found[0] for found in pattern.findall(article['text'])] for article in articles]
```

```python
file_lines[0]
```

## 25. テンプレートの抽出

```python
pattern = regex.compile(r'{{基礎情報([^{}]+|(?<rep>{{([^{}]+|(?&rep))*}}))*}}', re.MULTILINE)
templates = []
for article in articles:
    found = pattern.search(article['text'])
    if found is None:
        continue
    templates.append(found.group())
```

```python
templates[1]
```

```python
pattern = re.compile(r'\n\|\s*(?P<field>[^=]+?)\s*=\s*(?P<value>.+)')
infos = [pattern.findall(template) for template in templates]
```

```python
infos[1]
```

## 26. 強調マークアップの除去

```python
emphasis_removed = [[(field, re.sub(r"'''''|'''|''", r'', name)) for field, name in info] for info in infos]
```

```python
emphasis_removed[0]
```

## 27. 内部リンクの除去

```python
pattern = re.compile(r'\[\[(?P<title>[^(ファイル|File)].+?)(#.+)?(\|.+)?\]\]')
link_removed = [[(field, pattern.sub(r'\g<title>', name)) for field, name in info] for info in emphasis_removed]
```

```python
link_removed[0]
```

## 28. MediaWikiマークアップの除去

```python
pattern = re.compile(r'\[\[(ファイル|File).*\]\]')
file_removed = [[(field, pattern.sub(r'', name)) for field, name in info] for info in link_removed]
```

```python
file_removed[0]
```

## 29. 国旗画像のURLを取得する

```python
sess = requests.Session()
url = "https://en.wikipedia.org/w/api.php"
params = {
    "action": "query",
    "format": "json",
    "prop": "imageinfo",
    "iilimit": 50,
    "iiprop": "url",
}

pattern = re.compile(r' ')

country_to_flag_url = {}
for info in infos:
    info = dict(info)
    if '略名' not in info or '国旗画像' not in info:
        print(info)
        continue
    img_path = pattern.sub(r'_', info['国旗画像'])

    params["titles"] = 'File:{}'.format(img_path)
    request = sess.get(url=url, params=params)
    data = request.json()

    pages = list(data['query']['pages'].values())
    if not pages:
        print(info)
        continue

    img_url = pages[0]['imageinfo'][0]['descriptionurl']
    country_to_flag_url[info['略名']] = img_url
```
